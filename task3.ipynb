```python
import pandas as pd
```

```python
file_path = 'weather_data_extended.csv'
weather_data = pd.read_csv(file_path)
```

```python
# Research question are:-
# What is the distribution of temperature in the dataset?
# Is there a correlation between temperature and humidity?
# How does windspeed vary with temperature?
```

```python
# Use various visualization techniques to analyze the data and answer your research question:

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Distribution of Temperature

plt.figure(figsize=(10, 6))
sns.histplot(weather_data['Temperature (°C)'], kde=True)
plt.xlabel('Temperature')
plt.title('Distribution of Temperature')
plt.show()
```

```python
# Correlation between Temperature and Humidity

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature (°C)', y='Humidity (%)', data=weather_data)
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('Temperature vs Humidity')
plt.show()
```

```python
# Wind Speed vs Temperature

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature (°C)', y='Wind Speed (kph)', data=weather_data)
plt.xlabel('Temperature')
plt.ylabel('Wind Speed')
plt.title('Temperature vs Wind Speed')
plt.show()
```

```python
# plt.figure(figsize=(10, 6))

plt.plot(weather_data['Observation Time'], weather_data['Temperature (°C)'])
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Temperature Over Time')
plt.show()
```
